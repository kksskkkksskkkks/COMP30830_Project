```python
import pandas as pd

# 1. Load the dataset
df = pd.read_csv('input_data/final_merged_data.csv') 
# 2. Unit Conversion: Celsius -> Kelvin
df['temp'] = df['max_air_temperature_celsius'] + 273.15

# 3. Unit Conversion: Datetime String -> Millisecond Timestamp
# Convert to datetime format first, then to int64 nanoseconds, finally to milliseconds
df['last_reported_dt'] = pd.to_datetime(df['last_reported'])
df['timestamp_ms'] = (df['last_reported_dt'].view('int64') // 10**6)

# 4. Status Conversion: True -> OPEN
df['status'] = df['is_renting'].map({True: 'OPEN', False: 'CLOSED'})

# 5. Column Renaming (Directly updating the main DataFrame columns)
rename_dict = {
    'station_id': 'number',
    'capacity': 'bike_stands',
    'lon': 'lng',
    'num_bikes_available': 'available_bikes',
    'num_docks_available': 'available_bike_stands',
    'timestamp_ms': 'last_update', # Matches time in availability data
    'last_reported': 'dt',          # Matches raw datetime string in current data
    'max_relative_humidity_percent': 'humidity',
    'max_barometric_pressure_hpa': 'pressure'
}

df_cleaned = df.rename(columns=rename_dict)

# 6. Filter final required columns 
final_columns = [
    'number', 'name', 'address', 'lat', 'lng', 'bike_stands',          # Station attributes
    'available_bikes', 'available_bike_stands', 'status', 'last_update', # Availability attributes
    'dt', 'temp', 'humidity', 'pressure'                               # Weather attributes
]

# Check if these columns exist in the DataFrame and extract them
df_final = df_cleaned[[col for col in final_columns if col in df_cleaned.columns]]

# 7. Fill in missing constant fields 
df_final['contract_name'] = 'dublin'
df_final['banking'] = 0
df_final['bonus'] = 0

# Save the processed master table
df_final.to_csv('input_data/processed_training_data.csv', index=False)

print("Data processing complete: Columns renamed, temperature converted to Kelvin, and timestamps converted to milliseconds.")
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error, r2_score
import joblib
import pickle

# --- 1. Load Data ---
df = pd.read_csv("input_data/processed_training_data.csv")

# --- 2. Time Processing ---
df['timestamp'] = pd.to_datetime(df['dt'])

# --- 3. Feature Engineering ---
df['hour'] = df['timestamp'].dt.hour
df['day_of_week'] = df['timestamp'].dt.dayofweek

# Add optional weather features
if 'rain_1h' in df.columns:
    df['rain_1h'] = df['rain_1h'].fillna(0)
if 'snow_1h' in df.columns:
    df['snow_1h'] = df['snow_1h'].fillna(0)

# Clean up missing values
df.dropna(subset=['temp', 'humidity', 'pressure', 'available_bikes', 'bike_stands'], inplace=True)

# --- 4. Prepare Training Data ---
features = [
    'number',       # Station ID
    'bike_stands',  # Total station capacity
    'temp',         # Temperature
    'humidity',     # Humidity
    'pressure',     # Pressure
    'hour',         # Extracted hour
    'day_of_week'   # Extracted day of the week
]

# Prediction target
target = 'available_bikes' 

X = df[features]
y = df[target]

# Split the dataset
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# --- 5. Train Model ---
model = LinearRegression()
model.fit(X_train, y_train)

# --- 6. Evaluate and Save ---
y_pred = model.predict(X_test)
mae = mean_absolute_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

print("-" * 30)
print(f"LinearRegression model training completed! Prediction target: available bikes (available_bikes)")
print(f"Features used: {features}")
print(f"Mean Absolute Error (MAE): {mae:.2f}")
print(f"R² Score: {r2:.4f}")
print("-" * 30)

# Save the model
joblib.dump(model, "process_file/bike_availability_lr_model.joblib")
with open("process_file/bike_availability_lr_model.pkl", "wb") as f:
    pickle.dump(model, f)

print("The new model has been successfully saved.")
```

```python
import pandas as pd
import time
from sklearn.model_selection import train_test_split
from sklearn.neural_network import MLPRegressor       
from sklearn.preprocessing import StandardScaler      
from sklearn.pipeline import Pipeline                
from sklearn.metrics import mean_absolute_error, r2_score
import joblib


# --- 1. Load Data ---
print("Loading data...")
df = pd.read_csv("input_data/processed_training_data.csv")

# --- 2. Time Processing ---
df['timestamp'] = pd.to_datetime(df['dt'])

# --- 3. Feature Engineering ---
print("Performing feature engineering...")
# Extract time features
df['hour'] = df['timestamp'].dt.hour
df['day_of_week'] = df['timestamp'].dt.dayofweek

# Add new feature: distinguish between weekdays and weekends/holidays (0-4 are weekdays, 5-6 are weekends)
df['is_weekend'] = df['day_of_week'].apply(lambda x: 1 if x >= 5 else 0)

# Handle missing weather features (if any)
if 'rain_1h' in df.columns:
    df['rain_1h'] = df['rain_1h'].fillna(0)
if 'snow_1h' in df.columns:
    df['snow_1h'] = df['snow_1h'].fillna(0)

# Clean up invalid rows with missing values in core fields
df.dropna(subset=['temp', 'humidity', 'pressure', 'available_bikes', 'bike_stands'], inplace=True)

# Perform One-Hot Encoding on station numbers
# Because 'number' is a categorical variable, not a numerical magnitude. This step turns each station into a feature column (e.g., number_42, number_85...)
df = pd.get_dummies(df, columns=['number'], drop_first=False)

# --- 4. Prepare Training Data ---
print("Preparing model input data...")
# Define base features
base_features = [
    'bike_stands',  # Total station capacity
    'temp',         # Temperature
    'humidity',     # Humidity
    'pressure',     # Pressure
    'hour',         # Hour
    'day_of_week',  # Day of the week
    'is_weekend'    # Is weekend
]

# Add optional weather features
if 'wind_speed' in df.columns:
    base_features.append('wind_speed')
if 'rain_1h' in df.columns:
    base_features.append('rain_1h')

# Extract all station feature column names generated by One-Hot encoding
# pandas' get_dummies will add a prefix to the original column name, which defaults to the column name itself
station_features = [col for col in df.columns if col.startswith('number_')]

# Final list of all features used
features = base_features + station_features

target = 'available_bikes' # Target variable: number of available bikes

X = df[features]
y = df[target]

# Split the dataset
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)






print("-" * 30)
print("Building Neural Network (MLP) pipeline...")

# [Core Configuration]: Set up the pipeline, standardize first, then feed into the neural network
model_pipeline = Pipeline([
    ('scaler', StandardScaler()),  
    # Configure MLP model parameters
    ('mlp', MLPRegressor(
        hidden_layer_sizes=(128, 64, 32), # Hidden layer architecture: 3 layers, number of neurons decreases like a "funnel"
        activation='relu',                # Activation function: ReLU is the go-to for deep learning
        solver='adam',                    # Optimization algorithm: Adam usually converges the fastest
        alpha=0.001,                      # L2 regularization penalty coefficient to prevent overfitting
        batch_size=256,                   # Batch size: take 256 samples each time to update weights
        max_iter=500,                     # Maximum number of iterations (training epochs)
        early_stopping=True,              # [Crucial] Early stopping mechanism: if the model finds it can't learn anymore, it stops early to save time!
        random_state=42
    ))
])

print("Training the Neural Network model, please wait patiently...")
start_time = time.time()

# Start training
model_pipeline.fit(X_train, y_train)

end_time = time.time()
print(f"Training completed! Time taken: {end_time - start_time:.1f} seconds")

# --- Evaluate and Save ---
print("Evaluating the model...")
y_pred = model_pipeline.predict(X_test)
mae = mean_absolute_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

print("-" * 30)
print(f"MLP Neural Network Evaluation Results:")
print(f"Mean Absolute Error (MAE): {mae:.2f}")
print(f"R² Score: {r2:.4f}")
print(f"Actual iterations: {model_pipeline.named_steps['mlp'].n_iter_}") # See how many epochs it ran before early stopping
print("-" * 30)

# Save the entire pipeline
joblib.dump(model_pipeline, "output_model/bike_availability_mlp_pipeline.joblib")
print("MLP pipeline saved successfully!")
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_absolute_error, r2_score
import joblib
import pickle
import warnings
warnings.filterwarnings('ignore') 

# --- 1. Load Data ---
print("Loading data...")
df = pd.read_csv("input_data/processed_training_data.csv")

# --- 2. Time Processing ---
df['timestamp'] = pd.to_datetime(df['dt'])

# --- 3. Feature Engineering ---
print("Performing feature engineering...")
# Extract time features
df['hour'] = df['timestamp'].dt.hour
df['day_of_week'] = df['timestamp'].dt.dayofweek
df['is_weekend'] = df['day_of_week'].apply(lambda x: 1 if x >= 5 else 0)

# ================= Construct Lag Features =================
# Sort strictly by station ID and time first to prevent data mixing!
df = df.sort_values(by=['number', 'timestamp'])

# Group by and shift down one row to get the bike count of the "previous record" for that station
df['bikes_1h_ago'] = df.groupby('number')['available_bikes'].shift(1)

# Drop the null values in the first row generated by shift(1) (the first record of each station has no historical data)
df.dropna(subset=['bikes_1h_ago'], inplace=True)
# ==========================================================

# Handle missing weather features (if any)
if 'rain_1h' in df.columns:
    df['rain_1h'] = df['rain_1h'].fillna(0)
if 'snow_1h' in df.columns:
    df['snow_1h'] = df['snow_1h'].fillna(0)

# Clean up invalid rows with missing values in core fields
df.dropna(subset=['temp', 'humidity', 'pressure', 'available_bikes', 'bike_stands'], inplace=True)

# Perform One-Hot Encoding on station numbers
# This step will turn each station into a feature column (e.g., number_42, number_85...)
df = pd.get_dummies(df, columns=['number'], drop_first=False)

# --- 4. Prepare Training Data ---
print("Preparing model input data...")
# Define base features
base_features = [
    'bike_stands',  # Total station capacity
    'bikes_1h_ago', # Number of bikes in the previous hour
    'temp',         # Temperature
    'humidity',     # Humidity
    'pressure',     # Pressure
    'hour',         # Hour
    'day_of_week',  # Day of the week
    'is_weekend'    # Is weekend
]

# Add optional weather features
if 'wind_speed' in df.columns:
    base_features.append('wind_speed')
if 'rain_1h' in df.columns:
    base_features.append('rain_1h')

# Extract all station feature column names generated by One-Hot Encoding
station_features = [col for col in df.columns if col.startswith('number_')]

# Final list of all features used (base features + dummy variables for all stations)
features = base_features + station_features

target = 'available_bikes' # Target variable: available bikes

X = df[features]
y = df[target]

# Split the dataset
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# --- 5. Train Model ---
print("-" * 30)
print("Training the Random Forest model, please wait patiently...")

model = RandomForestRegressor(n_estimators=100, max_depth=15, n_jobs=-1, random_state=42)
model.fit(X_train, y_train)

# --- 6. Evaluate and Save ---
print("Evaluating the model...")
y_pred = model.predict(X_test)
mae = mean_absolute_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

print("-" * 30)
print(f"Training of Random Forest model with lag features complete! Prediction target: {target}")
print(f"Base features: {base_features}")
print(f"Number of station features: {len(station_features)}")
print(f"Total number of features: {len(features)}")
print(f"Mean Absolute Error (MAE): {mae:.2f}")
print(f"R² Score: {r2:.4f}")
print("-" * 30)

# Save the model 
print("Saving the model and feature list...")
joblib.dump(model, "process_file/bike_availability_rf2_model_final.joblib")

# The feature list must be saved! When predicting with new data in the future, it must be fed to the model in the exact order of this list
with open("process_file/bike_availability_rf2_model_final.pkl", "wb") as f:
    pickle.dump(features, f)

print("Model and feature configuration files have been successfully saved!")
```

```python
import pandas as pd
import time
from sklearn.model_selection import train_test_split
from sklearn.neural_network import MLPRegressor       
from sklearn.preprocessing import StandardScaler      
from sklearn.pipeline import Pipeline                 
from sklearn.metrics import mean_absolute_error, r2_score
import joblib


# --- 1. Load Data ---
print("Loading data...")
df = pd.read_csv("input_data/processed_training_data.csv")

# --- 2. Time Processing ---
df['timestamp'] = pd.to_datetime(df['dt'])

# --- 3. Feature Engineering ---
print("Performing feature engineering...")
df['hour'] = df['timestamp'].dt.hour
df['day_of_week'] = df['timestamp'].dt.dayofweek
df['is_weekend'] = df['day_of_week'].apply(lambda x: 1 if x >= 5 else 0)

# Handle missing weather features
if 'rain_1h' in df.columns:
    df['rain_1h'] = df['rain_1h'].fillna(0)
if 'snow_1h' in df.columns:
    df['snow_1h'] = df['snow_1h'].fillna(0)

# Clean invalid rows 
df.dropna(subset=['temp', 'humidity', 'pressure', 'available_bike_stands', 'bike_stands'], inplace=True)

# Perform One-Hot Encoding on station numbers
df = pd.get_dummies(df, columns=['number'], drop_first=False)

# --- 4. Prepare Training Data ---
print("Preparing model input data...")
base_features = [
    'bike_stands',   # Total station capacity
    'temp',          
    'humidity',      
    'pressure',      
    'hour',          
    'day_of_week',   
    'is_weekend'     
]

# Add optional weather features
if 'wind_speed' in df.columns:
    base_features.append('wind_speed')
if 'rain_1h' in df.columns:
    base_features.append('rain_1h')

station_features = [col for col in df.columns if col.startswith('number_')]
features = base_features + station_features

# Target variable: Number of available bike stands 
target = 'available_bike_stands' 

X = df[features]
y = df[target]

# Split the dataset
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)


print("-" * 30)
print("Building Neural Network (MLP) pipeline...")

# Set up pipeline: standardize first, then pass to neural network
model_pipeline = Pipeline([
    ('scaler', StandardScaler()),  
    # Configure MLP model parameters
    ('mlp', MLPRegressor(
        hidden_layer_sizes=(128, 64, 32), # Hidden layer architecture: 3 layers, neuron count decreases in a "funnel" shape
        activation='relu',                # Activation function: ReLU is the Swiss Army knife of deep learning
        solver='adam',                    # Optimization algorithm: Adam usually converges the fastest
        alpha=0.001,                      # L2 regularization penalty term to prevent overfitting
        batch_size=256,                   # Batch size: update weights using 256 samples each time
        max_iter=500,                     # Maximum number of iteration (training) rounds
        early_stopping=True,              # [Key] Early stopping mechanism: stops early to save time if the model stops improving!
        random_state=42
    ))
])

print("Training the Neural Network model, please wait patiently...")
start_time = time.time()

# Start training
model_pipeline.fit(X_train, y_train)

end_time = time.time()
print(f"Training completed! Time elapsed: {end_time - start_time:.1f} seconds")

# --- Evaluation and Saving ---
print("Evaluating the model...")
y_pred = model_pipeline.predict(X_test)
mae = mean_absolute_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

print("-" * 30)
print(f"MLP Neural Network Evaluation Results:")
print(f"Mean Absolute Error (MAE): {mae:.2f}")
print(f"R² Score: {r2:.4f}")
print(f"Actual iterations: {model_pipeline.named_steps['mlp'].n_iter_}") # See how many rounds it ran before early stopping
print("-" * 30)

# Save the entire pipeline
joblib.dump(model_pipeline, "output_model/bike_stands_mlp_pipeline.joblib")
print("MLP pipeline saved successfully!")
```
